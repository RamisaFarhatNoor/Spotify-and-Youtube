# Apply ML algorithm to predict the number of stream of a song on the basis of it's features only

### Preparing the features and the target variable

```python
stream_data = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo','Stream']
new_data_stream = data[stream_data]
val = new_data_stream.iloc[: , :-1 ]
target = new_data_stream.iloc[: , -1:]
```

### Checking Multicolinearity 

```python
corr_matrix = new_data_stream.corr()

# print the correlation matrix
print(corr_matrix)

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Show plot
plt.show()
```

## Observation: Some of the independent variables seem to be fairly dependent on another independent variable

## Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

value_scaled = scaler.fit_transform(val)


target_scaled = scaler.fit_transform(target)
```

## Selecting the Best Features 

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn import preprocessing
```

```python
selector = SelectKBest(score_func=f_regression, k=7)
value_new = selector.fit_transform(value_scaled, target_scaled)
```

```python
mask = selector.get_support() #list of booleans
new_features = [] # The list of your K best features

for bool, feature in zip(mask, val.columns):
    if bool:
        new_features.append(feature)

print(new_features)
```

## Dropping some features due to violation of multicolinearity and the unselected features 

```python
stream = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo']
# Assume that value_scaled is a NumPy array and cols_to_drop is a list of column names to drop
cols_to_drop = ['Key', 'Loudness','Speechiness', 'Valence', 'Tempo']

# Convert the NumPy array to a pandas DataFrame
df = pd.DataFrame(value_scaled, columns=stream)

# Use the .drop() method to remove the desired columns
df = df.drop(columns=cols_to_drop)

# Convert the pandas DataFrame back to a NumPy array
value_new = df.to_numpy()
value_new
```

## Training and testing data

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(value_new, target_scaled, test_size=0.2, random_state=42)
```

## Linear Regression and checking it's accuracy

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score



# Create a linear regression model
model = LinearRegression()

# Train the model on the training set
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the R-squared score
r2 = r2_score(y_test, y_pred)

print("R-squared score:", r2)
```

# Observation: The accuracy test is too low which indicates that the ML algorithm is unfit for this model

## Checking homoscedasticity of the model

```python
residuals = y_test - y_pred

# Plot residuals against predicted values
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

## Observation: Violates homoscedasticity and potrays heteroscedasticity. This is another indicator why Linear Regression shouldn't be used.

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
dtree = DecisionTreeRegressor()

dtree.fit(X_train , y_train)
y_pred = dtree.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

# Print the mean squared error
print("Mean squared error:", mse)
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python

# Create a random forest regressor object
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model with training data
rf_regressor.fit(X_train, y_train)

# Predict the output for test dataset
y_pred = rf_regressor.predict(X_test)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)

# Print the mean squared error
print("Mean squared error:", mse)
```

## SVR

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel = 'rbf')
svr.fit(X_train , y_train)
y_pred = svr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

print("Mean squared error:", mse)
```

## Observation: The lesser the MSE value is the better the model is. After applying 4 algorithms, it can be assumed that Random Forest Regressor is the best fit for this model and can predict the number of streams of a song  with their Danceability, Energy, Acousticness, Instrumentalness , Liveness better than other Regressors.
